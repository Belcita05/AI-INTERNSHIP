```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "C:/Users/DELL/Downloads/titanic.csv"
df = pd.read_csv(file_path)
df
```

```python
print("Missing values:\n", df.isnull().sum())
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
```

```python
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
df['Sex'] = le1.fit_transform(df['Sex'])

le2 = LabelEncoder()
df['Survived'] = le2.fit_transform(df['Survived'])
le3 = LabelEncoder()
df['Embarked'] = le3.fit_transform(df['Embarked'])
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_pred_log = log_reg.predict(X_test)
print("\nLogistic Regression Results")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log, target_names=["Died", "Survived"]))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred_log)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Died", "Survived"])
disp.plot()
plt.title("Confusion Matrix - Titanic Survival Prediction")
plt.show()
```
